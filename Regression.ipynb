## 1) Regression :


```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pylab as plt

# Données
X = torch.linspace(-2, 2, 200).view(-1, 1)
y = 2 * X + 1 + 0.2 * torch.randn_like(X)

# Modèle
model = nn.Sequential(
    nn.Linear(1, 16), nn.ReLU(),
    nn.Linear(16, 16), nn.ReLU(),
    nn.Linear(16, 1)
)

# Loss + optimizer
criterion = nn.MSELoss()
```

##### 1. Observer la loss:
On remarque que la loss diminue progressivement. Cela signfifie que le modele marche bien et qu'il apprend.

```python
losses = []

for epoch in range(100):
    optimizer.zero_grad()
    
    y_pred = model(X)
    loss = criterion(y_pred, y)
    
    loss.backward()
    optimizer.step()
    
    losses.append(loss.item())

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
    import matplotlib.pyplot as plt

plt.plot(losses)
plt.title("Evolution de la loss - Régression")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

##### 2. Comparaison des performances :
Régression
La loss (MSE) diminue progressivement
Le modèle approxime correctement la relation linéaire
Convergence stable et rapide

---->Bonne performance globale

Classification binaire
La loss (BCE) diminue mais plus lentement
Le modèle apprend une frontière non linéaire
Convergence moins rapide

---->Performance correcte mais problème plus difficile

Classification multi-classe
La loss (CrossEntropy) diminue rapidement
Le modèle sépare bien les différentes classes
Convergence rapide et stable

---->Très bonne performance

##### 3. Tester d'autres activations:


```python
model = nn.Sequential(
    nn.Linear(1, 16), nn.Tanh(),
    nn.Linear(16, 16), nn.Tanh(),
    nn.Linear(16, 1)
)

# Loss + optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Entraînement
for epoch in range(200):
    optimizer.zero_grad()
    
    y_pred = model(X)
    loss = criterion(y_pred, y)
    
    loss.backward()
    optimizer.step()
    
    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

print("Training terminé")
losses = []
for epoch in range(100):
    optimizer.zero_grad()
    y_pred = model(X)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
print(len(losses))  
plt.plot(losses)
plt.show()
```

```python
model = nn.Sequential(
    nn.Linear(1, 16), nn.Sigmoid(),
    nn.Linear(16, 16), nn.Sigmoid(),
    nn.Linear(16, 1)
)

# Loss + optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Entraînement
for epoch in range(200):
    optimizer.zero_grad()
    
    y_pred = model(X)
    loss = criterion(y_pred, y)
    
    loss.backward()
    optimizer.step()
    
    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

print("Training terminé")
losses = []
for epoch in range(100):
    optimizer.zero_grad()
    y_pred = model(X)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
print(len(losses))  
plt.plot(losses)
plt.show()
```

Avec ReLU, la loss diminue rapidement et atteint une valeur faible, ce qui indique une convergence rapide et efficace.
Avec Tanh, la convergence est correcte mais plus lente que ReLU.
Avec Sigmoid, la diminution de la loss est plus lente et la performance finale est moins bonne.

ReLU est la fonction d’activation la plus efficace, tandis que Sigmoid est moins adaptée .

##### 4. Modifier le learning rate:


```python
losses_lr1 = []

optimizer = optim.Adam(model.parameters(), lr=0.1)

for epoch in range(100):
    optimizer.zero_grad()
    
    y_pred = model(X)
    loss = criterion(y_pred, y)
    
    loss.backward()
    optimizer.step()

    losses_lr1.append(loss.item())

losses_lr2 = []

optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(100):
    optimizer.zero_grad()
    
    y_pred = model(X)
    loss = criterion(y_pred, y)
    
    loss.backward()
    optimizer.step()

    losses_lr2.append(loss.item())

losses_lr3 = []

optimizer = optim.Adam(model.parameters(), lr=0.001)

for epoch in range(100):
    optimizer.zero_grad()
    
    y_pred = model(X)
    loss = criterion(y_pred, y)
    
    loss.backward()
    optimizer.step()

    losses_lr3.append(loss.item())

import matplotlib.pyplot as plt

plt.plot(losses_lr1, label="lr=0.1")
plt.plot(losses_lr2, label="lr=0.01")
plt.plot(losses_lr3, label="lr=0.001")

plt.legend()
plt.title("Comparaison des learning rates")
plt.show()
```

Le learning rate influence fortement l’apprentissage du modèle.

Avec un learning rate élevé (0.1), la loss présente de fortes oscillations, ce qui montre une instabilité et une difficulté à converger.
Avec un learning rate intermédiaire (0.01), la loss diminue rapidement et de manière stable, ce qui indique un apprentissage efficace.
Avec un learning rate faible (0.001), la diminution de la loss est très lente, ce qui ralentit l’apprentissage.

Dans notre cas, 0.01 est le meilleur compromis.

##### 4. BCELoss vs BCEWithLogitsLoss:

La différence entre BCELoss et BCEWithLogitsLoss concerne l’utilisation de la fonction d’activation en sortie.

BCELoss nécessite que la sortie du modèle passe par une fonction Sigmoid afin d’obtenir une probabilité entre 0 et 1.
BCEWithLogitsLoss intègre directement la fonction Sigmoid dans le calcul de la loss, donc il ne faut pas ajouter de Sigmoid dans le modèle.
